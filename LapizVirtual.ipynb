```python
import cv2 
import numpy as np

cap = cv2.VideoCapture(0 , cv2.CAP_DSHOW)

celesteBajo = np.array([75, 185, 88], np.uint8)
celesteAlto = np.array([112, 255, 255], np.uint8)

# Colores para pintar
colorCeleste = (255,113,82)
colorAmarillo = (89,222,255)
colorGris = (148, 148, 148)
colorRosa = (128,0,255)
colorVerde = (0,255,36)
colorNegro =(0, 0, 0)
colorNaranja =(20, 117, 240)
colorVioleta = (134,1,175)
colorLimpiarPantalla = (29,112,246) # Solo se usará para el cuadro superior de 'Limpiar Pantalla'

# Grosor de línea recuadros superior izquierda (color a dibujar)
grosorCeleste = 3
grosorAmarillo = 3
grosorRosa = 3
grosorVerde = 3
grosorNegro = 3
grosorGris = 3
grosorNaranja = 3
grosorVioleta = 3

# Grosor de línea recuadros superior derecha (grosor del marcador para dibujar)
grosorPeque = 6
grosorMedio = 1
grosorGrande = 1

#--------------------- Variables para el marcador / lápiz virtual -------------------------
color = colorCeleste  # Color de entrada, y variable que asignará el color del marcador
grosor = 4 # Grosor que tendrá el marcador
#------------------------------------------------------------------------------------------




x1 = None
y1 = None
imAux = None



while True:
    
    ret,frame= cap.read()
    if ret == False: break
        
        
    frame = cv2.flip(frame,1) # flip para colocar el efecto espejo en la camara
    frameHSV = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV) # Transformamos la imagen a HSV 
    if imAux is None: imAux = np.zeros(frame.shape, dtype = np.uint8 ) # Creamos el lienzo de zeros del tamaño de nuestro frame
    
    #-------------------------------------------------------------- seccion superios --------------------------------------------------
    
    cv2.rectangle(frame,(0,0),(50,50),colorAmarillo,grosorAmarillo)# Creamos los rectangulos para pintar 
    cv2.rectangle(frame,(50,0),(100,50),colorRosa,grosorRosa)# Creamos los rectangulos para pintar if 
    cv2.rectangle(frame,(0,53),(50,100),colorGris,grosorGris)# Creamos los rectangulos para pintar 
    cv2.rectangle(frame,(100,0),(150,50),colorNegro,grosorNegro)# Creamos los rectangulos para pintar
    cv2.rectangle(frame,(150,0),(200,50),colorVerde,grosorVerde)# Creamos los rectangulos para pintar
    cv2.rectangle(frame,(200,0),(250,50),colorCeleste,grosorCeleste)# Creamos los rectangulos para pintar
    cv2.rectangle(frame,(0,103),(50,150),colorNaranja,grosorNaranja)# Creamos los rectangulos para pintar
    cv2.rectangle(frame,(0,153),(50,200),colorVioleta,grosorVioleta)# Creamos los rectangulos para pintar
    
    # Rectangulo central que limpiara la pantalla 
    
    cv2.rectangle(frame,(300,0),(450,50), (250,250,250),4)
    cv2.putText(frame,'Limpiar',(320,20),6,0.7, (0,0,0), 0 , cv2.LINE_AA) # colocamos el Texto
    cv2.putText(frame,'Pantalla',(340,40),6,0.7, (0,0,0), 0 , cv2.LINE_AA) # colocamos el Texto
    
    
    # Cuadro de dibujos en la parte superior derecha (grosor del marcador para dibujar)
    
    cv2.rectangle(frame, (490,0),(540,50),(0,0,0),grosorMedio)
    cv2.circle(frame,(515,25),3,(0,0,0),-1)
    cv2.rectangle(frame, (540,0),(590,50),(0,0,0),grosorMedio)
    cv2.circle(frame,(565,25),7,(0,0,0),-1)
    cv2.rectangle(frame, (590,0),(640,50),(0,0,0),grosorMedio)
    cv2.circle(frame,(615,25),11,(0,0,0),-1)
    
    # Dibujamos el lapiz 
    
    maskCeleste = cv2.inRange(frameHSV,celesteBajo,celesteAlto)
    #Realizaremos operaciones morfologicas Eliminando ruido
    
    maskCeleste = cv2.erode(maskCeleste,None,iterations = 1)
    maskCeleste = cv2.dilate(maskCeleste, None ,iterations = 2)
    
    #Realizaremos un suavisado para mejorar el ruido del lapiz 
    maskCeleste = cv2.medianBlur(maskCeleste,13)
    
    
    # Ahora buscamos los contornos 
    cnts,_ = cv2.findContours(maskCeleste,cv2.RETR_EXTERNAL,cv2.CHAIN_APPROX_SIMPLE)
    cnts = sorted(cnts, key=cv2.contourArea , reverse=True)[:1]
    
    for c in cnts:
        area = cv2.contourArea(c)
        if area > 1500:
            x,y2,w,h = cv2.boundingRect(c)
            #cv2.rectangle(frame(x,y2),(x+w,y2+h),(0,250,0),2)
            
            x2 = x + w//2
            
            if x1 is not None:
                if 0 < x2 < 50 and 0 < y2 < 50:
                    color = colorAmarillo # Color del lápiz/marcador virtual
                    grosorAmarillo = 6
                    grosorRosa = 2
                    grosorVerde = 2
                    grosorCeleste = 2
                if 50 < x2 < 100 and 0 < y2 < 50:
                    color = colorRosa # Color del lápiz/marcador virtual
                    grosorAmarillo = 2
                    grosorRosa = 6
                    grosorVerde = 2
                    grosorCeleste = 2
                if 100 < x2 < 150 and 0 < y2 < 50:
                    color = colorNegro # Color del lápiz/marcador virtual (100,0),(150,50)
                    grosorAmarillo = 2
                    grosorRosa = 2
                    grosorVerde = 6
                    grosorCeleste = 2
                if 150 < x2 < 200 and 0 < y2 < 50:
                    color = colorVerde # Color del lápiz/marcador virtual 
                    grosorAmarillo = 2
                    grosorRosa = 2
                    grosorVerde = 6
                    grosorCeleste = 2
                if 200 < x2 < 250 and 0 < y2 < 50:
                    color = colorCeleste # Color del lápiz/marcador virtual
                    grosorAmarillo = 2
                    grosorRosa = 2
                    grosorVerde = 2
                    grosorCeleste = 6
                if 0 < x2 < 50 and 103 < y2 < 150:
                    color = colorNaranja # Color del lápiz/marcador virtual (0,103),(50,150)
                    grosorAmarillo = 2
                    grosorRosa = 2
                    grosorVerde = 2
                    grosorCeleste = 6
                if 0 < x2 < 52 and 50 < y2 < 100:
                    color = colorGris # Color del lápiz/marcador virtual 
                    grosorAmarillo = 2
                    grosorRosa = 2
                    grosorVerde = 2
                    grosorCeleste = 6
                if 0 < x2 < 50 and 153 < y2 < 200:
                    color = colorVioleta # Color del lápiz/marcador virtual (0,153),(50,200)
                    grosorAmarillo = 2
                    grosorRosa = 2
                    grosorVerde = 2
                    grosorCeleste = 6
                if 490 < x2 < 540 and 0 < y2 < 50:
                    grosor = 3 # Grosor del lápiz/marcador virtual
                    grosorPeque = 6
                    grosorMedio = 1
                    grosorGrande = 1
                if 540 < x2 < 590 and 0 < y2 < 50:
                    grosor = 7 # Grosor del lápiz/marcador virtual
                    grosorPeque = 1
                    grosorMedio = 6
                    grosorGrande = 1
                if 590 < x2 < 640 and 0 < y2 < 50:
                    grosor = 11 # Grosor del lápiz/marcador virtual
                    grosorPeque = 1
                    grosorMedio = 1
                    grosorGrande = 6
                    
                if 300 < x2 < 450 and 0 < y2 < 50:
                    cv2.rectangle(frame,(300,0),(400,50),(255,255,255),3) # (300,0),(450,50)
                    cv2.putText(frame,'Limpiar',(320,20),6,0.7,(255,255,255),3,cv2.LINE_AA)
                    cv2.putText(frame,'pantalla',(320,40),6,0.7,(255,255,255),3,cv2.LINE_AA)
                    imAux = np.zeros(frame.shape,dtype=np.uint8)
                if 0 < y2 < 60 or 0 < y1 < 60 :
                    imAux = imAux
            
                else:
                    imAux = cv2.line(imAux,(x1,y1),(x2,y2),color,grosor)
            cv2.circle(frame,(x2,y2),grosor,color,5)
            x1= x2
            y1 = y2
            
        else:
            x1=None
            y1 = None
            
                    
    imAuxGray = cv2.cvtColor(imAux, cv2.COLOR_BGR2GRAY)          
    _, th = cv2.threshold(imAuxGray, 10 , 255 , cv2.THRESH_BINARY)
    thinv = cv2.bitwise_not(th)
    frame = cv2.bitwise_and(frame,frame, mask = thinv)
    frame = cv2.add(frame,imAux)
    

    cv2.imshow('frame', frame)
    #cv2.imshow('th', th)
    #cv2.imshow('maskRoja', maskCeleste)
    #cv2.imshow('thinv', thinv)
    #cv2.imshow('imAux', imAux)
    
    
    k = cv2.waitKey(1)
    if k == 27:
        break
        
cap.release()
cv2.destroyAllWindows()
```
